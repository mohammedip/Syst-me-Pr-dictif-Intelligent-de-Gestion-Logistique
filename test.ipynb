```python
from pyspark.sql import SparkSession

spark = SparkSession.Builder().getOrCreate()
df = spark.read.csv('app/data/DataCoSupplyChainDataset.csv' , header=True)
```

```python
df.show(5)
```

```python
from pyspark.sql.functions import *


df.filter(col('Late_delivery_risk')  == 1).groupBy('Order Region').count().show()
```

```python
df.filter(col('Late_delivery_risk') == 1).agg(avg(col('Sales'))).show()
```
